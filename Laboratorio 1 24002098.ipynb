# 1. <a id="division_1">GIT</a>

## 1.1. ¿Que es GIT?

Es un software de control de versiones. Su objetivo es llevar registros de las modificaciones realizadas a un proyecto.
Su eficiencia, confiabilidad y compatibilidad de versiones lo hacen el  mas utilizado en la actualidad.

No es un historial de versiones "lineal" donde se van guardando en una sola copia las modificaciones realizadas, sino que es un DVCS (sistema de control de versiones distribuido) donde cada usuario posee su propio historial de versiones alojado en su repositorio (el historia de versiones se visualiza a traves de ramificaciones)

Unas de las caracteristicas mas importantes

- **Rendimiento:** Permite a los usuarios hacer modificaciones en diferentes versiones y al final poder conciliar todos los cambios en las ramas principales.

- **Seguridad**: Conservar la integridad del codigo fuente. Todo esta protegido a través de una encriptación llamada "SHA1", para asegurar este proposito

- **Flexibilidad**: En diferentes aspectos: Utilización de varios flujos de trabajo. Adaptablidad tanto en pequeños como grandes proyectos. Compatibildiad con diversos sistemas y protocolos.

**RAMIFICACIONES** <br>
Git almacen las instantanes del trabajo realizado, es decir almacena todo, no solo las diferencias.
Al ejecturar el comando <code>git commit</code>, git realiza sumas de control de cada directorio y se guardan como objetos árbol en el repositorio Git. Despues se crea un objeto de confirmación con los datos y un apuntador al objeto árbol raiz del proyecto.

Una rama Git es simplemente un apuntador móvil apuntando a una de esas confirmaciones. La rama por defecto de Git es la rama master. Con la primera confirmación de cambios que realicemos, se] creará esta rama principal master apuntando a dicha confirmación.

**FLUJOS DE TRABAJO** <br>
Son acuerdos a los que se llegan para coordinar el trabajo de un grupo de personas.
<br>
Como se ha mencionado, existen diferentes flujos de trabajo que se pueden implementar en Git. Los mas comunes son:

1. **Git-Flow:** Es el flujo de trabajo más conocido. Es utilizado para proyectos que tienen entregables y ciclos bien definidos (por ejemplo versiones). Esta montado sobre dos ramas infinitas (Master y Development, para integración de cambios)) y varias ramas de apoyo: feature: nuevas funcionalidades (viene de development), hotfix: corrección de errores (viene de master), y release: preparación de nuevas versiones (viene de development y se fusiona con master y development).
2. **GitHub-Flow:** Es el utilizado en Github. Siempre con la misma dinamica de Git-Flow, se posee una rama master y las diferentes subramas. La diferencia radica en 2 puntos:
<ul>
    <li>Se trabaja directamente en subramas de master.</li>
    <li>La integración de cambios, es necesario abrir un "pull request" (solicitud de integración de cambios) para que otros usuarios puedan revisar y aprobar los cambios.</li>
3. **GitLab-Flow:** Similar a GitHub-Flow. Se añaden dos ramas
    <ul>
        <li>Ramas de entorno</li>
        <li>Ramas de versión</li>
4. **One Flow:** Cada nueva versión de producción está basada en la versión previa de producción. La mayor diferencia entre One Flow y Git Flow es que One Flow no tiene rama de desarrollo.

# 2. <a id="division_2">GITHUB</a>

## 2.1 ¿Que es GITHUB?

Es una plataforma que nos permite guardar repositorios de Git que podemos usar como servidores remotos y ejecutar algunos comandos de forma visual e interactiva (sin necesidad de comandos).<br>

En Github también puedes entrar a los proyectos de los demás y colaborar para mejorarlos.
Ofrece tambien una herramienta de revisión de codigo donde se pueden colocar anotaciones por si se quiere discutir sobre algo.<br>
Caracteristicas principales:
- **Esta formado por repositorios:** Son lugares virtuales alojados en la nube en donde los usuarios pueden almacenar cualquier tipo de archivo.
- **Alojamiento de código:** Permte alojar proyectos de codigo fuente de forma gratuita. Pueden ser proyectos publicos o privados.
- **Control de versiones:** Utilizando Git.
- **Colaboración:** Ofrece herramientas para facilitar la colaboración entre equipos. Gestion de ramas, pull requests, merge, revisar código.
- **Comunidad:** Posee una gran comunidad donde facilmente puedes encontrar ayuda y soporte y tambien colaborar en otros proyectos de código abierto. Algunas funciones dentro de la comunidad:
    <ul>
        <li>Crear Issues: Similar a los hilos en un foro común</li>
        <li>Debates (Discussions):  Pueden crearse sobre temas especificos y almacenar comentarios de otros usuarios. A diferencia de los issues, hay un sistema de votos para los comentarios y el usuario administrador puede marca una respuesta como "correcta".
        <li>Creacion de Wiki: Para dar mas detalles sobre un proyecto especifico sin añadir mas información dentro de las carpetas que alojan el codigo</li>

# 3. <a id='division_3'>MAGIC CELLS</a>

Es un sistema de comandos que provee de forma efectiva un lenguaje similar a la sintaxis de python y puede ser extendible por el usuario con nuevos comandos.

Los comandos se pueden dividir en dos tipos:
1. **Lineas magicas:** Estos comandos empiezan con un <code> % </code> y cuyos argumentos so
2. **Celdas magicas:** Estos utilizan dos porcentajes <code> % </code> y toman como argumentos todo lo que se puede alojar en la celda. Siempre se debe utlizar al inicio de la celda

Al utilizar <code> %lsmagic </code> se muestran todos los comandos disponibles. <br>
De forma similar <code>%quickref</code> muestra un mayor detalle de los comandos disponibles.

Algunos ejemplos <br>
<code>%timeit print("Hola") </code> Mide el tiempo de ejecución de la linea que se coloca despues del line magics <br>
<code>%%timeit (colocar codigo) </code> Mide el tiempo de ejecución de todo el codigo dentro de la celda

- Se pueden ejecutar codigos en otros lenguajes de programación. Ejemplos:
<code>%%Html</code>,
<code>%%JavaScript</code>,
<code>%%SparkSQL</code> <br>

- Se puede correr un archivo de python a traves de <code>%run "filename"</code>
- Listar variables definidas en el workbook: <code>%who</code>
- Utilizar comandos de la consola conda: <code>%conda</code>. Incluso se pueden instalar librerias <code>%conda install</code>
- Cargar un archivo externo a la celda (por ejemplo un .py que contenga funciones): <code>%load "filename.py"</code>
- Utilizar el debuggind de python (turn on / turn off): <code>%pdb</code>
- Ver el detalle técnico de los objetos: <code>%pinfo "variable"</code>
- Grabar el contenido de la celda en un archivo: <code>%%writefile "name.py" {codigo}</code>
